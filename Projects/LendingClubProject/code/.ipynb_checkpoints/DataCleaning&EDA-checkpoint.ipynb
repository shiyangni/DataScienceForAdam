```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

pd.set_option('display.max_columns', 200)
```

```python
sns.set()
```

```python
rawData = pd.read_excel("../data/LC_raw.xlsx")
n = len(rawData["Loan Status"])

removeCols = [column for column in rawData.columns if rawData[column].isnull().all()]
removeCols.extend([column for column in rawData.columns if rawData[column].count()<(n*(1-1))])
removeCols.extend([column for column in rawData.columns if rawData[column].nunique()<2])

rawData.drop(removeCols,axis=1,inplace=True)
DefaultDict = {"Charged Off":1,"Fully Paid":0}
rawData["Loan Status"].replace(DefaultDict,inplace=True)
#rawData["Interest rate"]=rawData["Interest Rate"].str.rstrip('%').astype('float') / 100.0

rawData.to_excel("LC_Cleaned.xlsx",index=False)
```

```python
df = pd.read_excel("../data/LC_Cleaned.xlsx")
```

```python
df_original = df

df["Funded Ratio"] = df["Amount Funded"]/df["Loan Amount"]
df[["Funded Ratio"]].describe()

## Proportion of 1 s.
(df["Funded Ratio"] == 1).sum()/df.shape[0]

df_notFullyFunded = df[df["Funded Ratio"] != 1]


df_notFullyFunded.shape[0]
(df_notFullyFunded["Funded Ratio"] < 0.6).sum()
sns.distplot(df_notFullyFunded["Funded Ratio"], rug=False, kde=False, hist_kws=dict(edgecolor="k", linewidth=1));
```
