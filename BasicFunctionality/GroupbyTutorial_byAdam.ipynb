## DataFrame.groupby()

Grouping data. Used in conjunction with either __aggregating, transforming or filtering the data.__

```python
# import the pandas library


ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
df
```

```python
# Viewing groups
df.groupby('Team').groups

# Multiple columns
grouped = df.groupby(["Team", "Year"])
grouped.groups
```

```python
## Iterating through groups. 
## Groupby transforms each data entry into a dictionary item, with groupby keys as 
## keys, and the rest as values.
## Iteration helps display the groupby object.
for name, group in grouped:
    name
    group

## But the easiest way would be to use describe.
grouped.describe()
```

```python
## Getting groups. Notice we must supply the correct datatype for keys.
grouped.get_group(("Devils", 2014))

## Or we can take a look through first()/last()
grouped.first()
grouped.last()
```

### Aggregating

```python
## Groupby object has the method agg. DataFrameGroupBy aggregates over all features,
## while SeriesGroupBy object(created from slicing from a DataFrameGroupBy object)
## aggregates over itself.
# A SeriesGroupBy
df.groupby("Team")["Points"]
df.groupby("Team")["Points"].agg([np.sum, np.mean, np.std])
# A DataFrameGroupBy
df.groupby("Team")
df.groupby("Team").agg([np.sum, np.mean, np.std])
```

### Transforming

<span style=color:red >Transformation and filtering </span> are rather confusing when used with group_by. We skip them.

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
df

grouped = df.groupby('Team')
score = lambda x: (x - x.mean()) / x.std()*10
print(grouped.transform(score))
```
